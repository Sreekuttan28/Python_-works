# Product Analysis

```python
import pandas as pd


data = {
    'ProductID': range(101, 111),
    'ProductName': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Category': [
        'Electronics', 'Electronics', 'Furniture', 'Furniture',
        'Clothing', 'Clothing', 'Grocery', 'Grocery',
        'Sports', 'Sports'
    ],
    'Price': [299.99, 199.99, 399.99, 499.99, 59.99, 49.99, 9.99, 14.99, 24.99, 19.99],
    'Stock': [50, 60, 30, 20, 100, 150, 200, 180, 90, 120],
    'QuantitySold': [150, 200, 120, 80, 300, 250, 400, 350, 220, 180]
}

df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Price'],df['Stock'])
plt.show()
```

## Adding Revenue column

```python
df['Revenue'] = df['Price'] * df['QuantitySold']
```

```python
df['StockValue'] = df['Price'] * df['Stock']
```

### Bar Chart of Revenue by Category

```python
category_revenue = df.groupby('Category')['Revenue'].sum()
category_revenue.plot(kind='bar', color='orange', title='Revenue by Category')
plt.ylabel('Total Revenue')
plt.show()
```
